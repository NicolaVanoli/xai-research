## Generating the Education attribute

```python
liv_dict = {"Big City Center":4,"Big City Suburbs":2.4,"Small City":3.2,"Rural":2.1}

def f_age_edu(age):
    if age > 30:
        res = 240
    else:
        res = 8*age
    return res

def f_liv_edu(liv):
    res = liv_dict[liv]*75
    return res

# function that ouputs the wieghted sum of age and liv_area values
def education(age,liv):
    res = (f_age_edu(age) + f_liv_edu(liv))/2
    return res
```

```python
import random
import numpy as np
age = [random.randrange(20,60) for i in range(100)]
age = np.array(age)
```

```python
bcc = np.full(shape=25, fill_value="Big City Center")
bcs = np.full(shape=25, fill_value="Big City Suburbs")
ru = np.full(shape=25, fill_value="Rural")
sc = np.full(shape=25, fill_value="Small City")
liv_area = np.concatenate([bcc,bcs,ru,sc])
```

```python
liv_area = np.expand_dims(liv_area, axis = 1)
age = np.expand_dims(age,axis = 1)
data = np.hstack([age,liv_area])
```

```python
data.shape
```

```python
import pandas as pd
df = pd.DataFrame(data,columns =['Age','Living Area'])
```

```python
df
```

```python
def score_to_cat(el):
    
        if el < 180:
            el = 'Middle School'
        elif el >= 180 and el < 240:
            el = 'High School'
        elif  el >= 240 and el < 274:
            el = 'University'
        elif el >= 274:
            el = 'PhD'
       
        return el
```

```python
edu = np.empty((100,1), dtype='object')

for index,row in df.iterrows():
    curr_age = row['Age']
    curr_liv = row['Living Area']
    res = education(int(curr_age),curr_liv)
    noise = np.random.normal(0,8)
    noise_res = res + noise
    #adding noise
    
    res = score_to_cat(noise_res)
    edu[index] = res
df['Education'] = edu
```
